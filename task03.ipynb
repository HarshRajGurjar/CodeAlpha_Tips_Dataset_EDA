# Task 3.1:  Setup and Data Loading

```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=sns.load_dataset("tips")
```

```python
df.head()
```

# Task 3.2: Visualization of Total Bill vs. Tip Amount

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='total_bill', y='tip', data=df, hue='sex', style='smoker',s=80)
plt.title('Relationship Between Total Bill and Tip Amount')
plt.xlabel('Total Bill ')
plt.ylabel('Tip Amount ')
plt.legend(title='Customer Info')
plt.grid(True)
plt.savefig("Plot1.png")
plt.show()
```

# Task 3.3:  Visualization of Tips by Day of the Week

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='day', y='tip', data=df, palette='viridis', order=['Thur', 'Fri', 'Sat', 'Sun'])
plt.title('Distribution of Tips by Day of the Week')
plt.xlabel('Day of the Week')
plt.ylabel('Tip Amount ($)')
plt.savefig("plot2.png")
plt.show()
```

# Task 3.4: Visualization of Tip Percentage by Time of Day

```python
df['tip_percentage'] = (df['tip'] / df['total_bill']) * 100
```

```python
plt.figure(figsize=(8, 6))
sns.barplot(x='time', y='tip_percentage', data=df, palette='coolwarm')
plt.title('Average Tip Percentage by Time of Day')
plt.xlabel('Time of Day')
plt.ylabel('Average Tip Percentage (%)')
plt.savefig("plot3.png")
plt.show()
```

1. Scatter Plot: Total Bill vs. Tip Amount

1.1 Data Story: This plot tells a clear story about the relationship between what a customer spends and how much they tip. You can show that there's a strong, positive correlation: as the total bill increases, the tip amount tends to increase as well.

1.2 Portfolio Impact: By using hue='sex' and style='smoker', you demonstrate your ability to add multiple layers of information to a single chart. This shows you can create a complex, insightful visualization that answers several questions at once.

2. Box Plot: Tips by Day of the Week

2.1 Data Story: This visualization reveals a clear pattern in tipping behavior over the course of a week. You can see that tips are distributed differently on weekends (Saturday and Sunday) compared to weekdays (Thursday and Friday). This insight could be used to inform staffing or marketing decisions.

2.2 Portfolio Impact: A box plot is a perfect way to compare distributions. It shows that you understand how to use the right chart for the right type of data, which is a key skill for a data analyst.

3. Bar Plot: Average Tip Percentage by Time of Day
 
3.1 Data Story: This plot provides a direct comparison of a key business metric: tip percentage during lunch versus dinner. It can reveal if customers are more generous at certain times, which is a powerful insight for revenue optimization.

3.2 Portfolio Impact: By first calculating a new variable (tip_percentage) and then visualizing it, you show that you can not only plot data but also manipulate and engineer new features to get a deeper understanding of the data.
